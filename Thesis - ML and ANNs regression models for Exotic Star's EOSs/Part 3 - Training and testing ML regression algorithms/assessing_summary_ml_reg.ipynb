```python
# Saving MSE results
assess_ML(star_type="NS",mag_reg="enrg",x_data_types=["16X_rwsh","32X_rwsh"],EOS_type="lin").save_metric_results("mse") # results for energy density on center ε_c prediction
```

## **1.2 Printing the metrics results in PrettyTable format**

```python
# MSLE results
print(">Energy on center ε_c prediction (MSLE results):")
assess_ML(star_type="NS",mag_reg="enrg",x_data_types=["16X_rwsh","32X_rwsh"],EOS_type="lin").print_metric_results("msle")
# MSE results
print("\n>Energy on center ε_c prediction (MSE results):")
assess_ML(star_type="NS",mag_reg="enrg",x_data_types=["16X_rwsh","32X_rwsh"],EOS_type="lin").print_metric_results("mse")
```

## **1.3 Making groups of bar plots**

```python
# Initialzing the figure and axis
fig_gbar_enrg_NS,axis_gbar_enrg_NS = plt.subplots(2,1,figsize=(10,8))

# Defining the bar colors
bar_clrs_msle = ["cadetblue","goldenrod"]
bar_clrs_mse = ["mediumseagreen","salmon"]

# MSLE results
assess_ML(star_type="NS",mag_reg="enrg",x_data_types=["16X_rwsh","32X_rwsh"],EOS_type="lin").make_bar_plots("msle",axis_gbar_enrg_NS[0],bar_colors=bar_clrs_msle)
# MSE results
assess_ML(star_type="NS",mag_reg="enrg",x_data_types=["16X_rwsh","32X_rwsh"],EOS_type="lin").make_bar_plots("mse",axis_gbar_enrg_NS[1],bar_colors=bar_clrs_mse)

fig_gbar_enrg_NS.tight_layout()
```
